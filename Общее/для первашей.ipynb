**Сессия**

1. Зимняя и летняя сессия - т.е. 2 шт
2. Каникулы перед зимней сессией государственные, потом 2 недели после сессии (выходите на учебу в начале февраля) <br>
Летом каникулы только после сессии, выходите на учебу в сентябре

**Учебная система\как закрыться**

1. Система 100-бальная =>  максимум можно получить 100 баллов за предмет<br>оценки<br>3 - 60 баллов<br>4 - 71<br>5 - 85<br> (сама сумма баллов имеет значение только если вы боретесь за бюджетное место)<br><br>
2. У вас есть обязательные работы: РК, КР, ДЗ (без них вы не получите допуск\зачет)<br>РК - пишете на паре теория + задачи<br>КР - только задачи, пишете на паре<br>ДЗ - делаете дома, но сдаете<br>Если вы сдаете невовремя без уважительной причины (болезнь и подобное), то вы все равно обязаны сдать обязательные мероприятия, однако получаете за них минимум в любом случае(т.к. система балльная вы из этих контрольных мероприятий получаете баллы. Для того, чтобы сдать работу - необходимо набрать по ней минимум) <br><br>
3. Если вы зяглянете в учебный план, то увидите интересные буквы (экз\зчт и т.д.) - это формат предмета<br>
ЭКЗ - экзамен (за контрольные мероприятия вы набираете максимум 70 баллов, далее на экзе добираете оставшиеся 30 баллов. Минимум за экзамен - 18 баллов, иначе отправляетесь на пересдачу)<br>
ЗЧТ - зачет (работа только в течение семестра, оценка не выставляется (в зачетке напишут зачет и все))<br>
ДЗЧТ - диф зачет (аналогично зачету, однако по нему уже идет оценка)<br><br>
КУР\КУП - курсовая работа\курсовой проект (пар по предмету нету (буквально дома делаете), работаете самостоятельно, выставляется оценка)<br>
РЭКЗ - распределенный экзамен (не стоит пугаться. Вы работаете в течение семестра, вам озвучвивают оценку, если вас оценка не устраивает, то вы можете пойти на экзамен, иначе - выставляется по факту текущих баллов)<br>
<br>
Примечание: препод имеет право отклоняться от балльной системы, поэтому слушайте, что он говорит и выполняйте его требования
