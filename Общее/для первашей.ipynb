Примечание: хотите узнать больше о преподе? Посмотрите информацию о нем на студизбе

**Сессия**

1. Зимняя и летняя сессия - т.е. 2 шт
2. Каникулы перед зимней сессией государственные, потом 2 недели после сессии (выходите на учебу в начале февраля) <br>
Летом каникулы только после сессии, выходите на учебу в сентябре

**Учебная система\как закрыться**

1. Система 100-бальная =>  максимум можно получить 100 баллов за предмет<br>оценки<br>3 - 60 баллов<br>4 - 71<br>5 - 85<br> (сама сумма баллов имеет значение только если вы боретесь за бюджетное место)<br>К слову, для зачета необходимо 60 баллов<br><br>
2. У вас есть обязательные работы: РК, КР, ДЗ (без них вы не получите допуск\зачет)<br>РК - пишете на паре теория + задачи<br>КР - только задачи, пишете на паре<br>ДЗ - делаете дома, но сдаете<br>Если вы сдаете невовремя без уважительной причины (болезнь и подобное), то вы все равно обязаны сдать обязательные мероприятия, однако получаете за них минимум в любом случае(т.к. система балльная вы из этих контрольных мероприятий получаете баллы. Для того, чтобы сдать работу - необходимо набрать по ней минимум) <br><br>
3. Если вы зяглянете в учебный план, то увидите интересные буквы (экз\зчт и т.д.) - это формат предмета<br>
ЭКЗ - экзамен (за контрольные мероприятия вы набираете максимум 70 баллов, далее на экзе добираете оставшиеся 30 баллов. Минимум за экзамен - 18 баллов, иначе отправляетесь на пересдачу)<br>
ЗЧТ - зачет (работа только в течение семестра, оценка не выставляется (в зачетке напишут зачет и все))<br>
ДЗЧТ - диф зачет (аналогично зачету, однако по нему уже идет оценка)<br><br>
КУР\КУП - курсовая работа\курсовой проект (пар по предмету нету (буквально дома делаете), работаете самостоятельно, выставляется оценка)<br>
РЭКЗ - распределенный экзамен (не стоит пугаться. Вы работаете в течение семестра, вам озвучвивают оценку, если вас оценка не устраивает, то вы можете пойти на экзамен, иначе - выставляется по факту текущих баллов)<br>
<br>
Примечание: препод имеет право отклоняться от балльной системы, поэтому слушайте, что он говорит и выполняйте его требования<br><br>
4. По предмету у вас могут быть следующие форматы пар:<br>
лекции - сидите всем потоком, слушаете, о чем препод глаголит (у шестилеток есть аж три потока, поздравляю)<br>
потоки шестилеток:
- поток иу5 (всем группы, включая бакалвров и гуимц)<br>
- поток шестилеток с вашей кафедры (перваши 2025 года соответственно две группы поток)<br>
- поток шестилеток иу5 + иу6 (будет на предметах некоторых, начиная со второго курса, например матвед 3 сем, физика 4 сем)<br><br>
семинары - сидите только своей группой (аналогичен обычному уроку в школе) <br>
лабы - сидите своей группой, делаете\сдаете лабы \ЛАБЫ ЛУЧШЕ НЕ СКИПАТЬ\

**Физра**

С физрой шутки плохи, ходить на нее надо в любом случае<br>
Обязательно 25 посещений, сдача нормативов + 60 баллов (т.к. зачетный предмет)<br>
В сентябре вы получите паспорта здоровья, которые вы обязаны продлевать раз в год<br>БЕЗ ПАСПОРТА ЗДОРОВЬЯ НЕПРОСРОЧЕННОГО ПРЕПОДАВАТЕЛЬ НЕ ИМЕЕТ ПРАВА ДОПУСТИТЬ ВАС К ЗАНЯТИЮ<br>На первом курсе вы можете податься в сборную, после 2 семестра - записаться на секцию<br>В паспорте здоровья врач вам выставит группу здоровья:
1. основная, подготовительная - особо ничем не отличаются, человек может записываться в секции, сдавать нормативы и поступать на военку
2. спецмед - лайт физра, нет минималки по сдаче нормативов, как говорится "спецмеду дают баллы за участие, а не за сдачу"
3. АФК - человек не имеет физической нагрузки вообще, у них отдельно от всех проходят пары, на которых они занимаются обычно рефератами и подобными штуками
<br><br>
Скажу сразу, группу здоровья, при желании также можно поменять, обратитесь в студенческую поликлиннику
<br>ВАЖНО: на военку и секции принимают только основную и подготовительную группы

**Где у меня пара**

Быстрый гайд<br>
1. посмотрите номер кабинета в расписании (обычно там 3 цифры + буква)<br>
- 1 цифра - этаж
- остальные цифры - номер кабинета
- буква - корпус
2. букв достаточно много<br>
- гз - буква ю (южное крыло), отстутствие буквы северное крыло
- улк - буква л
- химлаб - буква х
- конгресс холл - буква к
- мнок - буква т

Пример: 337 (3 этаж, гз), 567к (5 этаж, конгресс холл), 234ю (2 этаж, гз), 102.5 (1 этаж, все также гз) и т.п.
